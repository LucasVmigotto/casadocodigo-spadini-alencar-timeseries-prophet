```python
import plotly
from plotly import offline as py
from plotly import graph_objects as go
from plotly.offline import (
    plot,
    iplot
)
import cufflinks as cf

cf.go_offline()

plotly.offline.init_notebook_mode(connected=True)

from plotly import io as pio

# Usar esta linha se o ambiente for o Google Colab
# pio.renderers.default = 'colab'

from prophet.plot import (
    plot_plotly,
    plot_components_plotly,
    plot_seasonality_plotly
)

from prophet.serialize import model_from_json
from json import load

with open('./data/processed/climate_change_br.json') as file_ref:
    model = model_from_json(load(file_ref))
    file_ref.close()

predictions = model.predict()

predictions.head()
```

## Gráfico interativo da previsão

```python
fig = plot_plotly(model, predictions, figsize=(1150, 750))

py.iplot(fig)
```

```python
fig = plot_plotly(model, predictions, trend=True, figsize=(1150, 750))

py.iplot(fig)
```

## Gráfico interativo componentes

```python
fig = plot_components_plotly(model, predictions, figsize=(1150, 175))

py.iplot(fig)
```

## Gráfico interativo da sazionalidade

```python
fig = plot_seasonality_plotly(model, 'yearly', figsize=(1150, 275))

py.iplot(fig)
```
